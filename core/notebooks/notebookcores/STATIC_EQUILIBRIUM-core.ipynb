```python
s = Solver()
F = []
r = []

FA = list(Reals('FAx FAy'))
FB = list(Reals('FBx FBy'))
FP = list(Reals('FPx FPy'))
FW = list(Reals('FWx FWy'))

rA = list(Reals('rAx rAy'))
rB = list(Reals('rBx rBy'))
rP = list(Reals('rPx rPy'))
rW = list(Reals('rWx rWy'))

F += [FA, FW, FP, FB]
r += [rA, rW, rP, rB]

s.add(FW[0] == 0)
s.add(FW[1] == -200)
s.add(FP[0] == 0)
s.add(FP[1] == -500)

s.add(ry == 0 for _, ry in r)
s.add(rA[0] == 0)
s.add(rW[0] == 2)
s.add(rP[0] == 3)
s.add(rB[0] == 4)

Fx = [x for x, _ in F]
Fy = [y for _, y in F]
s.add(sum(Fx) == 0)
s.add(sum(Fy) == 0)

tau = [cross(ri, Fi) for ri, Fi in zip(r, F)]
s.add(sum(tau) == 0)

print(s.check())
print(s.model())
```
