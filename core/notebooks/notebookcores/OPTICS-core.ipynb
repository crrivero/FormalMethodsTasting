```python
# d_i > 0 => real image
# d_i < 0 => virtual image
# h_i > 0 => erect image
# h_i < 0 => inverted image
# f > 0 => concave mirror
# f < 0 => convex mirror

s = SimpleSolver()

# focal length, distance image, distance object
f, d_i, d_o = Reals('f d_i d_o')
# s.add(1/f == 1/d_i + 1/d_o)
s.add(d_i*d_o == f*d_i + f*d_o)

# magnification, height image, height object
m, h_i, h_o = Reals('m h_i h_o')
s.add(m == h_i/h_o)
s.add(m == -d_i/d_o)

# radius of curved mirror
R = Real('R')
s.add(R == 2*f)

s.add(d_o == 6.0)
s.add(d_i == -1.0)
s.check()
s.solve_for([f])
```
