## Mate Matching

Consider a set of males and females within a population of a certain species. Not every male and mate with every female due to genetics, blood type, etc. We can model the relationship which individuals can mate using the following graph, where left verticies represent males and right verticies represent females. If two vertices are connected by an edge, that means that the two individuals can mate.



```python
males = [ 'M1', 'M2', 'M3' ]
females = [ 'F1', 'F2', 'F3' ]

compatability = [
    ( 'M1', 'F1' ), # M1 and F1 are compatible
    ( 'M1', 'F2' ),
    ( 'M2', 'F1' ),
    ( 'M2', 'F3' ),
    ( 'M3', 'F2' ),
]
draw_bipartite_graph( males, females, compatability )
```

We want to pair each male and female within the population so that each male is only paired to one female, and each female is only paired to one male, i.e., we want to **match** the males and females.

Formally, a **matching** is a collection of edges where no two vertices share an edge. We want to find a matching where each vertex on the right is part of an edge in the matching.

Let's see how to do that using Z3.

<!-- For the sake of completeness, we now provide some formal definitions. A bipartite graph  -->
<!-- Formally, given two disjoint sets of vertices $V_1$ and $V_2$ in a graph, a **matching** is a set of edges $M$ such that: (1) no two edges in $M$ share a vertex, and (2) each edge in $M$ has one vertex in $V_1$ and one vertex in $V_2$. -->

```python
# M1 shares an edge with F1, we will represent that edge with the variable M1F1:
M1F1 = Bool( 'M1F1' )

# We follow the same naming convention for our other variables

M1F2 = Bool( 'M1F2' )
M2F1 = Bool( 'M2F1' )
M2F3 = Bool( 'M2F3' )
M3F2 = Bool( 'M3F2' )
```

#### Step 2: Initialize the solver

```python
s = Solver()
```

#### Step 3: Write constraints

```python
# Each left vertex (male) can have at most one edge adjacent to it in the matching
s.add( M1F1 + M1F2 <= 1 )
s.add( M2F1 + M2F3 <= 1 )
# Note that we do not have to add the constraint "M3F2 <= 1" since Bob_Talk
# is a boolean variable and its value is always at most 1 when converted to
# an integer

# Each right vertex (female) can have at most one edge adjacent to it in the matching
s.add( M1F1 + M2F1 <= 1 )
s.add( M1F2 + M3F2 <= 1 )

# Now we need to add a constraint that ensures all of the right vertices are
# part of an edge in the matching. Since there are three tasks, we can do this
# by asking for the sum of all edges going into the females be at least three:

s.add( M1F1 + M1F2 + M2F1 + M2F3 + M3F2 >= 3 )

# Let's see what our constraints look like
showSolver( s )
```

#### Step 4: Check if a solution exists

```python
print( s.check() )
```

#### Step 5: View solution

```python
solution = s.model()
print(solution)
```
